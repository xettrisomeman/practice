```python
import pandas as pd
import matplotlib.pyplot as plt

from spacy.lang.ne.stop_words import STOP_WORDS
import re
import string
```

```python
laxmi = open('mehendi.txt' , 'r' , encoding='utf-8')
```

```python
text = laxmi.read()
```

```python
laxmi.close()
```

```python
text
```

```python
punct = list(string.punctuation)
```

```python
def change_into_tokens(text):
    text = text.replace('\ufeff' , ' ')
    text = re.sub('\s+' , ' ' , text)
    text = text.replace('\n' , ' ').replace('।' , ' ').replace('____' , '')
    text = text.replace('॥' , ' ')
    text = re.sub('\d+' , ' ',text)
    text = text.strip().split(' ')
    
    cleaned_text= []
    
    for token in text:
        if token not in STOP_WORDS and token not in punct and token!='':
            cleaned_text.append(token)
    
    return cleaned_text
```

```python
change_into_tokens("\ufeff[१/ \nप्रथम \nसुन सुन झिल्की \nक्या, नगवेली \nउषा परी \nहिमाल करी। \n[२] आँखा चडी \nसुनको पङ्खी ॥ \n[३] फर फर गर्दै \nहावा हम्की। गिरिशिर टेकी \nझिलमिल गम्की॥ \nसुरपुर मुस्की \nबास्ना फुस्की। मगमग हावा \nचल्दछ लस्की॥ \nआधा ब्यूँझी \nयूँगा झस्की । सरसिज फुस्की ____ मुसमुस मुस्की॥ \nहाँस्छे उषा \nलाली छी। कुसुम छ सुन्दर \nमोती लुर्की॥ \nधके चली \nसुन के फुली। झलझल झल्का \nनाला टल्की ॥ \nसपना मेरो \nसुनको फेरो। घुम्दै ब्यूँझयो \nनभ वन घेरो ॥ \n[९] रुनुमुनु छन्छन् \nअणु अणु बन्छन् । \nम्हेन्दु :१ \nआशा पखेरु \nनव स्वर लिन्छन् । \n[१०] यी हुन् \'म्हेन्दु ")
```

```python
#laxmi cleaned text

laxmi_cleaned_text = change_into_tokens(text)
```

```python
#create wordcloud

laxmi_wordcloud = ' '.join(laxmi_cleaned_text)
```

```python
from wordcloud import WordCloud
```

```python
font='F://anaconda//Lib//site-packages//wordcloud//Lohit-Devanagari.ttf'

fig , ax = plt.subplots(figsize=(50,50))
wordcloud = WordCloud(font_path=font,max_font_size=50,max_words=20,collocations=False , background_color='white').generate(laxmi_wordcloud)

ax.imshow(wordcloud)
ax.axis('off')
plt.show()
```
